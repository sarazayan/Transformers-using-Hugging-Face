```python
from transformers import pipeline 
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import textwrap
from sklearn.metrics import roc_auc_score,f1_score,confusion_matrix
```

```python
classifier=pipeline("zero-shot-classification")
```

```python
classifier("this is a great movie",candidate_labels=["positive","negative"])
```

```python
#load dataset
df=pd.read_csv('/content/sample_data/bbc_text_cls (1).csv')
```

```python
#check length of dataset
len(df)
```

```python
#samples of dataframe
df.sample(frac=1).head()
```

```python
#cast the label to set and then to a list 
labels=list(set(df['labels']))
labels
```

```python
#print a random document
print(textwrap.fill(df.iloc[1024]['text']))
```

```python
df.iloc[1024]['labels']
```

```python
classifier(df.iloc[1024]['text'],candidate_labels=labels)
```
