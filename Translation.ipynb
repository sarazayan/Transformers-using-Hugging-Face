```python
#load data and store in a dataset suitable for our needs aka dictionary 
eng2spa={}
for line in open('/content/spa-eng/spa-eng/spa.txt'):
    line=line.rstrip()
    eng,spa=line.split("\t")
    if eng not in eng2spa:
        eng2spa[eng]=[]
    eng2spa[eng].append(spa)
      
```

```python
#import a function from nltk to compute bleu score 
from nltk.translate.bleu_score import sentence_bleu,SmoothingFunction
from nltk.tokenize import RegexpTokenizer
from nltk import word_tokenize, pos_tag, WordNetLemmatizer
```

```python
#blue metric works with tokens rather than a row text
tokenizer=RegexpTokenizer(r'\w+'),#removes punctuation
```

```python
tokenizer  
```

```python
tokens=tokenizer.tokenize('¿Qué me cuentas?'.lower())
sentence_bleu([tokens],tokens)
```

```python
sentence_bleu([['hi']],['hi'])
```

```python
#add a smoothing function 3shan elwarning ellih 3elaka b 3adad elgrams da 
smoother=SmoothingFunction()
sentence_bleu(['hi'],'hi',smoothing_function=smoother.method4)
```

```python
sentence_bleu(['hi there'.split()],'hi there'.split())
```

```python
#we can even pass integers 
sentence_bleu([[1,2,3,4]],[1,2,3,4])
```

```python
#pretokenize our targets for later use
eng2spa_tokens={}
for eng,spa_list in eng2spa.items():
  spa_list_tokens=[]
  for text in spa_list:
    tokens=tokenizer.tokenize(text.lower())
    spa_list_tokens.append(tokens)
  eng2spa_tokens[eng]=spa_list_tokens 
```
