```python
from transformers import AutoTokenizer
```

```python
checkpoint='distilbert-base-uncased'
tokenizer=AutoTokenizer.from_pretrained(checkpoint)
```

```python
#test the tokenizer on a subsample of the sentences
tokenized_sentences=tokenizer(raw_datasets['train'][0:3]['sentence'])
from pprint import pprint
pprint(tokenized_sentences)
#only attention mask and input_ids because this is only for bert not distilbert
```

```python
def tokenize_fn(batch):
  return tokenizer(batch['sentence'],truncation=True)
```

```python
#map the tokenizer_fn 3shan fiha truncation=true to each of the samples in the dataset we da elhanb3ato ba3d keda leltrainer
tokenized_datasets=raw_datasets.map(tokenize_fn,batched=True)
```

```python
from transformers import TrainingArguments
```

```python
#define training args objects 
training_args=TrainingArguments(
    'my_trainer',
    evaluation_strategy='epoch',#to evalute the model
    save_strategy='epoch',#save model at each training step
    num_train_epochs=1
)
```

```python
from transformers import AutoModelForSequenceClassification
```

```python
#load the pretrained model
model=AutoModelForSequenceClassification.from_pretrained(
    checkpoint,num_labels=2
)
```

```python
#save model parameters before training
params_before=[]
for name,p in model.named_parameters():
  params_before.append(p.detach().cpu().numpy())
```

```python
from transformers import Trainer
```

```python
from datasets import load_metric
```

```python
metric=load_metric("glue","sst2")
```

```python
metric.compute(predictions=[1,0,1],references=[1,0,1])
```

```python
#define compute metrics function 
def compute_metrics(logits_and_labels):
  logits,labels=logits_and_labels
  predictions=np.argmax(logits,axis=-1)
  return metric.compute(predictions=predictions,references=labels)

```

```python
trainer=Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets['validation'],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics

)
```

```python
trainer.train()
```

```python
trainer.save_model('my_saved_model')
```
